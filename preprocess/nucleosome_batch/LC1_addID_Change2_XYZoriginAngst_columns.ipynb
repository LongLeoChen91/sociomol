```python
# -*- coding: utf-8 -*-
"""
Add sequential particle IDs to a STAR file.
"""

import starfile

# ===== Configuration =====
INPUT_STAR  = "nucleosome_top3_tomos.star"
OUTPUT_STAR = "ID_" + INPUT_STAR

# ===== Step 1: Read the STAR file =====
df = starfile.read(INPUT_STAR)

# ===== Step 2: Add sequential IDs =====
df["rlnTomoParticleId"] = range(1, len(df) + 1)

# ===== Step 3: Save =====
starfile.write(df, OUTPUT_STAR, overwrite=True)

print(f"✔ Added 'rlnTomoParticleId' (1–{len(df)})")
print(f"✔ Output saved to: {OUTPUT_STAR}")
```

```python
import starfile
import pandas as pd

input_star = "ID_nucleosome_top3_tomos.star"
output_star = "ID_nucleosome_top3_tomos_with_origin.star"

rename_map = {
    "rlnOriginX": "rlnOriginXAngst",
    "rlnOriginY": "rlnOriginYAngst",
    "rlnOriginZ": "rlnOriginZAngst",
}

origin_cols = [
    "rlnOriginXAngst",
    "rlnOriginYAngst",
    "rlnOriginZAngst",
]

def process_df(df):
    # Step 1: rename if old columns exist
    df = df.rename(columns=rename_map)

    # Step 2: ensure all origin columns exist
    for col in origin_cols:
        if col not in df.columns:
            df[col] = 0.0  # use float, not int

    return df

# ===== Read =====
star_data = starfile.read(input_star)

# ===== Process =====
if isinstance(star_data, pd.DataFrame):
    df = process_df(star_data)
    starfile.write({"particles": df}, output_star, overwrite=True)

elif isinstance(star_data, dict):
    for block_name, df in star_data.items():
        if isinstance(df, pd.DataFrame):
            star_data[block_name] = process_df(df)

    starfile.write(star_data, output_star, overwrite=True)

print(f"✔ Finished. Output written to: {output_star}")
```
