```python
import starfile
import pandas as pd

input_star = "R3_ID_Manual_1.star"
output_star = "R3_ID_Manual_1_with_origin.star"

rename_map = {
    "rlnOriginX": "rlnOriginXAngst",
    "rlnOriginY": "rlnOriginYAngst",
    "rlnOriginZ": "rlnOriginZAngst",
}

star_data = starfile.read(input_star)

if isinstance(star_data, pd.DataFrame):
    df = star_data.rename(columns=rename_map)
    starfile.write({"particles": df}, output_star, overwrite=True)

elif isinstance(star_data, dict):
    for block_name, df in star_data.items():
        if isinstance(df, pd.DataFrame):
            star_data[block_name] = df.rename(columns=rename_map)
    
    # ✅ write the full structure
    starfile.write(star_data, output_star, overwrite=True)

print(f"Finished. Output written to: {output_star}")
```

```python
# -*- coding: utf-8 -*-

import starfile
import glob

PIXEL_SIZE = 8  # Å per pixel

# find all matching files
star_files = [
    "arm1_0_R3_ID_Manual_1_with_origin.star",
    "arm1_1_R3_ID_Manual_1_with_origin.star",
    "arm2_0_R3_ID_Manual_1_with_origin.star",
    "arm2_1_R3_ID_Manual_1_with_origin.star",
]

for input_star in star_files:
    output_star = "Reset_" + input_star

    df = starfile.read(input_star)

    # adjust coordinates
    df["rlnCoordinateX"] -= df["rlnOriginXAngst"] / PIXEL_SIZE
    df["rlnCoordinateY"] -= df["rlnOriginYAngst"] / PIXEL_SIZE
    df["rlnCoordinateZ"] -= df["rlnOriginZAngst"] / PIXEL_SIZE

    # reset origins
    df["rlnOriginXAngst"] = 0.0
    df["rlnOriginYAngst"] = 0.0
    df["rlnOriginZAngst"] = 0.0

    starfile.write(df, output_star, overwrite=True)
    print(f"✔ {input_star} → {output_star}")
```
